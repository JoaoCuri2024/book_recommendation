```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
livros=pd.read_csv('BX_Books.csv',sep=';',encoding='latin1')
livros.head()
```

```python
user=pd.read_csv('BX-Users.csv',sep=';',encoding='latin1')
user.head()
```

```python
nota=pd.read_csv('BX-Book-Ratings.csv',sep=';',encoding='latin1')
nota.head()
```

```python
livros=livros[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
livros.rename(columns={'Book-Title':'titulo','Book-Author':'autor','Year-Of-Publication':'ano','Publisher':'editora'},inplace=True)
user.rename(columns={'User-ID':'usuario','Location':'local','Age':'idade'},inplace=True)
nota.rename(columns={'User-ID':'usuario','Book-Rating':'nota'},inplace=True)
```

```python
nota['usuario'].value_counts()
```

```python
x=nota['usuario'].value_counts()>200
```

```python
y=x[x].index
print(y.shape)
```

```python
nota=nota[nota['usuario'].isin(y)]
```

```python
rating_with_books=nota.merge(livros,on='ISBN')
rating_with_books.head()
```

```python
number_rating=rating_with_books.groupby('titulo')['nota'].count().reset_index()
number_rating=number_rating.sort_index(ascending=True)
```

```python
number_rating
```

```python
number_rating.rename(columns={'nota':'number_of_rating'}, inplace=True)
```

```python
final_rating= rating_with_books.merge(number_rating,on='titulo')
```

```python
final_rating.head(100)
```

ficar com os livros que foram avaliados mais de 50x

```python
final_rating=final_rating[final_rating['number_of_rating']>=50]
```

```python
final_rating.drop_duplicates(['usuario','titulo'],inplace=True)
```

```python
book_pivot=final_rating.pivot_table(columns='usuario',index='titulo',values='nota')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

Novas predições

```python
#1984
distances,suggestions=model.kneighbors(book_pivot.iloc[0,:].values.reshape(1,-1))
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
import numpy as np

# Selecionando um ponto de referência, por exemplo, o primeiro ponto no conjunto de dados
distances, indices = model.kneighbors(book_sparse[0], n_neighbors=5)

# Visualizando as distâncias dos 5 vizinhos mais próximos
print(f"Distâncias dos vizinhos mais próximos: {distances}")

```

```python

# Calculando a distância média para esses vizinhos
avg_distance = np.mean(distances)
print(f"Distância média: {avg_distance}")
```
